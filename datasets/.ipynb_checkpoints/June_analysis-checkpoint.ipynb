```python
session3 = pd.read_csv('session-4356-report-6_1_2022.csv')
session3.head()
```

```python
# Cleaning the column names
new_cols = [x.lower().replace(' ', '_').replace('-', '_') for x in session3.columns]
```

```python
session3.head(3)
```

```python
session3['p1:_investigate_a_dataset'].value_counts().sort_values(ascending=False)
```

```python
session3['p1:_investigate_a_dataset'].value_counts(normalize=True).sort_values(ascending=False).plot(kind='pie');
```

### For those who haven't submitted

```python
unsubs = session3.query('`p1:_investigate_a_dataset` == "Unsubmitted"')
```

```python
print(len(unsubs))
unsubs.head(3)
```

```python
unsubs_data = unsubs[['first_name', 'last_name', 'email']]
unsubs_data.head(3)
```

```python
unsubs_data.to_csv('p1_unsubs.csv', index=False)
```
