lets grab the names, email and attendance%, P1 and P2 completion rates

```python
df = att_12th_jul[['First Name', 'Last Name', 'Email', 
                   'P1: Investigate a Dataset', 
                   'P2: Wrangle and Analyze Data',
                   '% Attendance', 'Last Session Attended']]
df.head()
```

```python
attendance_15 = df[df['% Attendance']<15]
attendance_1530 = df[(df['% Attendance']>=15) & (df['% Attendance']<30)]
attendance_3045 = df[(df['% Attendance']>=30) & (df['% Attendance']<45)]
attendance_4560 = df[(df['% Attendance']>=45) & (df['% Attendance']<60)]
attendance_6080 = df[(df['% Attendance']>=60) & (df['% Attendance']<80)]
attendance_80plus = df[df['% Attendance']>=80]
print('done!')
```

```python
print(f'< 15:({len(attendance_15)}) \
\n15 - 30:({len(attendance_1530)}) \
\n30 - 45:({len(attendance_3045)}) \
\n45 - 60:({len(attendance_4560)}) \
\n60 - 80:({len(attendance_6080)}) \
\n>= 80: ({len(attendance_80plus)})')
```
