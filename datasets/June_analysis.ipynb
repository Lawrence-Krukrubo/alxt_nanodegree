```python
import pandas as pd
pd.set_option('max_columns', None)
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
print('All Imported!')
```

```python
os.listdir()
```

```python
def read_csv(str):
    df = pd.read_csv(str)
    return df
```

```python
def clean_cols(df):
    # Cleaning the column names
    new_cols = [x.lower().replace(' ', '_').replace('-', '_') for x in df.columns]
    df.columns = new_cols
    print('Columns cleaned!')
```

```python
def find_unsubs(df, col):
    unsubs = df.query(f'`{col}` == "Unsubmitted"')
    unsubs = unsubs[['first_name', 'last_name', 'email']]
    return unsubs
```

```python
def find_ungrades(df, col):
    ungrades = df.query(f'`{col}` == "Ungradeable"')
    ungrades = ungrades[['first_name', 'last_name', 'email']]
    return ungrades
```

```python
def find_incomps(df, col):
    incomps = df.query(f'`{col}` == "Incomplete"')
    incomps = incomps[['first_name', 'last_name', 'email']]
    return incomps
```

```python
# Cleaning the column names

clean_cols(session3)
```
