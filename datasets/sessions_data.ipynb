Check Attendance...

```python
pd.set_option('max_rows',None)
att16May = pd.read_csv('attendance_condensed16thMay.csv')
att16May.head()
```

```python
len(att16May)
```

```python
att16May.columns = ['full_name', 'email', 'duration_mins', 'is_guest']
```

```python
att16May['full_name'] = att16May.full_name.str.lower()
```

```python
att16May.index = att16May['full_name']
```

```python
att16May.drop('full_name', axis=1, inplace=True)
```

```python
att16May.head()
```

```python
# For first 10 names

def check_attendance(df, arr):
    assert min(df.duration_mins) >= 1
    present, absent = {}, {}
    
    for name in arr:
        if name in df.index:
            present[name] = 'Present'
        else:
            absent[name] = 'Absent'
    return present, absent
```

```python
df_16may.head(3)
```

```python
allStudents_16thMay = list(df_16may['First Name']+' '+df_16may['Last Name'])
allStudents_16thMay = [val.lower() for val in allStudents_16thMay]
```

```python
present, absent = check_attendance(att16May, allStudents_16thMay)
```

```python
print(f'Out of {len(allStudents_16thMay)}\n{len(present)} were present on 16th May\n{len(absent)} were absent.')
```
